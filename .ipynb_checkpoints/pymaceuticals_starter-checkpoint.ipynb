```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Read the mouse data and the study results
mouse1= pd.read_csv('dataset/Mouse_metadata.csv')
study1= pd.read_csv('dataset/Study_results.csv')

# Combine the data into a single DataFrame
combinedf = pd.merge(mouse1, study1, how='inner', on='Mouse ID') 

# Display the data table for preview
combinedf
```

```python
print(f"rows in mouse dataset: {len(mouse1)}")
print(f"rows in study dataset: {len(study1)}")
print(len(combinedf))
combinedf.to_csv('dataset/combinedf.csv', index=False)
```

```python
# Checking the number of mice.
len(combinedf['Mouse ID'].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mice = combinedf[combinedf.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]
duplicate_mice_ids = duplicate_mice['Mouse ID'].unique()
duplicate_mice_ids
```

```python
# Optional: Get all the data for the duplicate mouse ID.
g989df = combinedf[combinedf['Mouse ID'] == 'g989']
g989df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = combinedf.drop(combinedf[combinedf['Mouse ID']=='g989'].index)
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_df['Mouse ID'].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.
#summary statistics, Group By Drug Regimen and calulate stats
summary_stats_df = (clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)']
                    .agg(['mean', 'median', 'var', 'std', 'sem'])
                    .round(3)
                    .rename(columns={
                        'mean':'mean tumor volume',
                        'median':'median tumor volume',
                        'var':'variance tumor volume',
                        'std':'std dev tumor volume',
                        'sem':'sem tumor volume'}))
summary_stats_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_stats_df = (clean_df.groupby('Drug Regimen')[['Tumor Volume (mm3)']]
                    .agg(['mean', 'median', 'var', 'std', 'sem'])
                    .round(3)
                    .rename(columns={
                        'mean':'mean tumor volume',
                        'median':'median tumor volume',
                        'var':'variance tumor volume',
                        'std':'std dev tumor volume',
                        'sem':'sem tumor volume'}))

summary_stats_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
new_df = clean_df.groupby('Mouse ID')['Timepoint'].max().reset_index()
merged1 = new_df.merge(clean_df, on=['Mouse ID', 'Timepoint'], how='inner')

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged1
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_volumes = []

# Calculate the IQR and quantitatively determine if there are any potential outliers.
# Locate the rows which contain mice on each drug and get the tumor volumes
   
for drug in treatments:
    drug_volumes = merged1.loc[merged1['Drug Regimen']==drug, 'Tumor Volume (mm3)']
    tumor_volumes.append(drug_volumes)

for drug, drug_volumes in zip(treatments, tumor_volumes):
    Q1 = drug_volumes.quantile(0.25)
    Q3 = drug_volumes.quantile(0.75)

    IQR = Q3-Q1

    # Determine outliers using upper and lower bounds
    lower_bound = Q1-(1.5*IQR)
    upper_bound = Q3+(1.5*IQR)
    
    outliers = drug_volumes[ (drug_volumes<lower_bound) | (drug_volumes>upper_bound) ]

    #show the results
    print(f"{drug} IQR is:{IQR}")
    print(f"Outlier for {drug}: {outliers}\n")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

plt.boxplot(tumor_volumes, tick_labels=treatments, patch_artist=True, 
            flierprops=dict(marker='o', color='red', markersize=4))

plt.xlabel('Drug Regimen')
plt.ylabel('Tumor Volume')
plt.title('Distribution of Tumor Volume by Treatment Group')
plt.tight_layout()
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

capomulin_rats = clean_df[clean_df['Drug Regimen']=='Capomulin']
average_capomulin = capomulin_rats.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
weight_capomulin = capomulin_rats.groupby('Mouse ID')['Weight (g)'].mean()

plt.figure(figsize=(8,6))
plt.scatter(weight_capomulin, average_capomulin)
plt.xlabel('Average Mice Weight')
plt.ylabel('Average Observed Tumor Volume')
plt.title('Capomulin: Mice Weight vs Average Observed Tumor Volume')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

corr_coeff = weight_capomulin.corr(average_capomulin)
print(corr_coeff)
```

```python
slope, intercept, r_value, p_value, std_err = st.linregress(weight_capomulin, average_capomulin)
print(f"slope = {slope}")
print(f"intercept = {intercept}")
```

```python
# Plot the linear regression model on top of the previous scatter plot.

plt.figure(figsize=(8,6))
plt.scatter(weight_capomulin, average_capomulin, label='Data Points')
plt.plot(weight_capomulin, slope*weight_capomulin+intercept, color='red', label='Linear Regression')
plt.xlabel('Average Mice Weight')
plt.ylabel('Average Observed Tumor Volume')
plt.title('Capomulin: Mice Weight vs Average Observed Tumor Volume')
plt.legend()
plt.show()
```
